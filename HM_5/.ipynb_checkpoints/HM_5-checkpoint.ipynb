## *Условие* 

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):

1. p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
2. s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ.
3. l – list – команда, которая выведет список всех документов в формате
passport "2207 876234" "Василий Гупкин"
4. as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. Предусмотрите случай, когда пользователь добавляет полку, которая уже существует.
5. d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;
6. m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;
7. a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
print(documents)
```

```python
input_user = input("""
    Введидите команду из списка ниже
    p - кому принадлежит документ
    s – полка, на которой лежит документ
    l – список всех документов
    as – добавление новой полки
    d – удаление документа
    m – перемещение документа на другую полку
    a – добавление нового документа
    """)

if input_user == "p":
    def func_1():    
        doc_number = input("Введите номер документа= ")
        for data in documents:
            if doc_number == data['number']:
                return 'Документ принадлежит '+ data['name']
        return 'Документ не найден!'
    print(func_1())

elif input_user == "s":
    def func_2():
        doc_number = input("Введите номер документа= ")
        for key, value in directories.items():
            if doc_number in value:
                return 'Полка №' + key
        return 'Документ не найден!'

    print(func_2())

elif input_user == "l":
    for doc in documents:
        print(f'{doc["type"]} "{doc["number"]}" "{doc["name"]}" ')

elif input_user == "as":
    def func_3():
        shelf_number = input("Введите номер полки= ")
        for shelf in directories.keys():
            if shelf_number in shelf:
                return 'Такая полка уже существует'
            else:
                directories[shelf_number] = []
                return directories
    print(func_3())

elif input_user == "d":
    def func_4():
        doc_delete = input("Введите номер документа, который хотите удалить= ")
        i = -1
        for data_1 in documents:
            i += 1
            if doc_delete == data_1['number']:
                del(documents[i])
                print(documents)
        for data in directories.values():
                if doc_delete in data:
                    a = data.index(doc_delete)
                    del(data[a])
                    return directories
        return 'Документ не найден'
    print(func_4())

if input_user == "m":
    doc_num = input("Введите номер документа, который хотите перенести = ")
    if doc_num not in str(directories.values()):
        print('Документ не найден')
    else:
        shelf_num = input("Введите номер полки, куда перенести документ = ")
        if shelf_num not in (str(directories.keys())):
            print('Полка не найдена. Введите другу полку')
        else:
            def func_5():
                new_list = []
                for key, value in directories.items():
                    if doc_num in value:
                        a = value.index(doc_num)
                        new_list = value[a]
                        del(value[a])
                        directories[shelf_num] = value+[new_list]
                        return directories
            print(func_5())

elif input_user == "a":
    doc_type = input('Введите тип документа ')
    doc_number = input('Введите номер документа ')
    doc_person = input('Введите ответственное лицо ')
    shelf_choice = input('На какую полку поставить документ ')
    if shelf_choice not in str(directories.keys()):
        print('Такой полки не существует. Выберите существуюшую полку')
    else:
        result = {'type':doc_type, 'number':doc_number, 'name':doc_person}
        documents.append(result)
        print(documents)
        x = directories[shelf_choice]
        directories[shelf_choice] = x + [doc_number]
        print(directories)
else:
    print('Команда не найдена')
```
